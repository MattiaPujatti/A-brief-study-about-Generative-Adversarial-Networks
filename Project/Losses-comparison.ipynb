```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def minimax(x):
    return 0.5*np.log(1-x) 
```

```python
def non_saturating(x):
    return -0.5*np.log(x)
```

```python
def max_likelihood(x):
    return -0.5*np.exp(np.log(x/(1-x)))
```

```python
def least_squares(x):
    return np.square(1-x)
```

```python
eps = 1e-5
x = np.linspace(0 + eps, 1 - eps, 10000)

fig, ax = plt.subplots()
ax.plot(x, minimax(x), label="Minimax")
ax.plot(x, non_saturating(x), label="Non-saturating heuristic", color="red")
ax.plot(x, max_likelihood(x), label="Maximum likelihood")
ax.plot(x, least_squares(x), label="Least squares")

ax.set_ylim(-5,5)
ax.legend(loc="lower left")
ax.set_xlabel("D(G(z))")
ax.set_ylabel("$J^{(D)}$")
ax.set_title("Comparison of generator's cost functions")
```
